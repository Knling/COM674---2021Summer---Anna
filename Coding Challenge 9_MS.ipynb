### The question:
I want to collect data from Twitter about two types of vaccine, Pfizer and Moderna, and see:
* Are there any differences in the lengths of these tweets between two types of vaccine?
* Is the number of retweets associated with the number of favorites in these tweets about vaccine? If so, does this association depend on the type of vaccine? 

```python
#Authorization

import tweepy
import json
from twitter_authentication import CONSUMER_KEY, CONSUMER_SECRET, ACCESS_TOKEN, ACCESS_TOKEN_SECRET

auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
#get 1000 original tweets about Pfizer and save them in a .json file

results = []
for tweet in tweepy.Cursor(api.search, 
                           q='Pfizer -filter:retweets', # only get the original tweets
                           tweet_mode = 'extended',
                           count=200).items(1000): 
    results.append(tweet._json)
    print(tweet.user.screen_name + "\t" + str(tweet.created_at) + "\t" + tweet.full_text)

with open('Pfizer_tweets.json', 'w') as f:
    json.dump(results, f)    
```

```python
#get 1000 original tweets about Moderna and save them in a .json file

results = []
for tweet in tweepy.Cursor(api.search, 
                           q='Moderna -filter:retweets', # only get the original tweets
                           tweet_mode = 'extended',
                           count=200).items(1000): 
    results.append(tweet._json)
    print(tweet.user.screen_name + "\t" + str(tweet.created_at) + "\t" + tweet.full_text)

with open('Moderna_tweets.json', 'w') as f:
    json.dump(results, f)
```

```python
with open('Pfizer_tweets.json', 'r') as f:
    Pfizer_tweets = json.load(f)
with open('Moderna_tweets.json', 'r') as f:
    Moderna_tweets = json.load(f)
```

```python
#Filter and save the JSON data into a CSV file

import csv
with open('cleaned_vac_data.csv', 'w', 
          encoding='UTF-8',
          newline='') as fn:
    f = csv.writer(fn)
    f.writerow(['created_at',
                'tweet_text',
                'retweets',
                'favorites',
                'tweet_length',
                'vac_type'
               ])
    for tweet in Pfizer_tweets:
        f.writerow([tweet['created_at'], 
                    tweet['full_text'],
                    tweet['retweet_count'],
                    tweet['favorite_count'],
                    len(tweet['full_text']),
                    "Pfizer"
                   ])
    for tweet in Moderna_tweets:
        f.writerow([tweet['created_at'], 
                    tweet['full_text'],
                    tweet['retweet_count'],
                    tweet['favorite_count'],
                    len(tweet['full_text']),
                    "Moderna"
                   ])
```

```python
#import data from csv

import pandas as pd
import seaborn as sns
import numpy as np

df = pd.read_csv('./cleaned_vac_data.csv')
df
```

```python
#visualize the tweet length by types of vaccine

sns.boxplot(y='tweet_length', x= 'vac_type', data = df);
```

```python
#conduct an independent t-test

from scipy.stats import ttest_ind
ttest_ind(df.tweet_length[df.vac_type == 'Pfizer'],df.tweet_length[df.vac_type == 'Moderna'])
```

```python
#What are the tweets that have received most retweets?

df.sort_values(['retweets'],ascending=False).groupby('vac_type').head(3)
```

```python
#What are the tweets that have received most favorites?

df.sort_values(['favorites'],ascending=False).groupby('vac_type').head(3)
```

```python
#visualize the relation between # of retweets and # of favoriates by type of vaccine

sns.set(style="ticks") 
p=sns.jointplot(x=np.log(df.retweets + 1),
              y=np.log(df.favorites + 1),
              hue ='vac_type',
              data=df);
p.set_axis_labels('Retweets (log)','Favorites (log)');
```
